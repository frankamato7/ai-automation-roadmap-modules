### Task 5 - Create a new column
#### Call it "salary_level", with thresholds

```python
import numpy as np

conditions = [
    (df['salary'] < 70000),
    (df['salary'] >= 70000) & (df['salary'] <= 90000),
    (df['salary'] > 90000)
]

choices = [
    "Low",
    "Medium",
    "High"
    ]

df['salary_level'] = np.select(conditions,choices,default='other')
df
```
